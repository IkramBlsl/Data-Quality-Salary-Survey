```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Charger le fichier Excel dans un DataFrame
file = 'Ask A Manager Salary Survey 2021 (Responses).xlsx'  # Assure-toi de spécifier le bon chemin d'accès
```

```python
nouveaux_noms_colonnes = ['Timestamp', 'age', 'industry','job','complement_job','annual_salary','aditional_salary','currency','other_currency','additional_context','country','state','city','work_year','year_field','educational_level','gender','race']  # Remplace avec tes nouveaux noms de colonnes
```

```python
donnees = pd.read_excel(file, header=None, names=nouveaux_noms_colonnes)
```

```python
donnees.shape
```

```python
# Supprimer la première ligne
donnees = donnees.iloc[1:]
```

```python
#Réinitialiser les index après la suppression
donnees.reset_index(drop=True, inplace=True)
```

```python
# Compter le nombre de personnes dans chaque tranche d'âge
nombre_personnes_par_age = donnees['age'].value_counts()
```

```python
# Affichage graphique du nombre de personnes dans chaque tranche d'âge (histogramme)
plt.figure(figsize=(8, 4))
donnees['age'].value_counts().plot(kind='bar', color='skyblue')
plt.xlabel('Tranche d\'âge')
plt.ylabel('Nombre de personnes')
plt.title('Répartition du nombre de personnes par tranche d\'âge')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Afficher les valeurs uniques dans la colonne 'job'
valeurs_uniques_job = donnees['job'].unique()
print("Valeurs uniques dans la colonne 'job' :")
for job in valeurs_uniques_job:
    print(job)
```

```python
# Compter le nombre de personnes dans chaque genre
nombre_personnes_par_genre = donnees['gender'].value_counts()
```

```python
# Créer un graphique à barres pour représenter le nombre de personnes par genre
plt.figure(figsize=(8, 6))
nombre_personnes_par_genre.plot(kind='bar', color='skyblue')
plt.xlabel('Genre')
plt.ylabel('Nombre de personnes')
plt.title('Répartition du nombre de personnes par genre')
plt.xticks(rotation=8)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(nombre_personnes_par_genre, labels=None , autopct='%1.1f%%', startangle=140)
plt.legend(nombre_personnes_par_genre.index, loc='right', bbox_to_anchor=(1.5, 0.5))
plt.title('Répartition du nombre de personnes par genre')
plt.axis('equal')  
plt.tight_layout()
plt.show()
```

```python
# Afficher les valeurs uniques dans la colonne 'currency'
Curr=donnees['currency'].unique()
Curr
```

```python
#compter les occurences de chaque niveau d'éducation
pourcentages_education = donnees['educational_level'].value_counts(normalize=True) * 100
```

```python
# Créer un graphique à barres pour représenter les pourcentages des niveaux d'éducation
plt.figure(figsize=(8, 6))
pourcentages_education.plot(kind='bar', color='skyblue')
plt.xlabel('Niveau d\'éducation')
plt.ylabel('Pourcentage')
plt.title('Pourcentages des niveaux d\'éducation')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#créer un tableau croisé pour compter les occurences de chaque niveau d'éducation pour genre
tableau_croise_educ_gender = pd.crosstab(donnees['educational_level'], donnees['gender'])
```

```python
tableau_croise_educ_gender
```

```python
# Créer un graphique à barres groupées pour représenter le niveau d'éducation par genre
plt.figure(figsize=(8, 6))
tableau_croise_educ_gender.plot(kind='bar', color=['skyblue', 'salmon'], alpha=0.8)
plt.xlabel('Niveau d\'éducation')
plt.ylabel('Nombre de personnes')
plt.title('Niveau d\'éducation par genre')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Genre')
plt.tight_layout()
plt.show()
```

```python
# Créer un tableau croisé pour compter les occurrences de chaque race par genre
tableau_croise_race_gender = pd.crosstab(donnees['race'], donnees['gender'])
```

```python
# Afficher le tableau croisé (optionnel, pour vérification)
print(tableau_croise_race_gender)
```

```python
# Créer un graphique à barres groupées pour représenter les races par genre
plt.figure(figsize=(14, 8))  # Augmenter la taille de la figure
tableau_croise_race_gender.plot(kind='bar', color=['skyblue', 'salmon'], alpha=0.8)
plt.xlabel('Race')
plt.ylabel('Nombre de personnes')
plt.title('Répartition des races par genre')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Genre')
plt.tight_layout()
plt.show()
```

```python
# Créer un tableau croisé pour compter les occurrences de chaque race par niveau d'éducation
tableau_croise_race_education = pd.crosstab(donnees['race'], donnees['educational_level'])
```

```python
# Afficher le tableau croisé (optionnel, pour vérification)
print(tableau_croise_race_education)
```

```python
# Créer un graphique à barres groupées pour représenter les races par niveau d'éducation
plt.figure(figsize=(14, 8))
tableau_croise_race_education.plot(kind='bar', stacked=True, alpha=0.8)
plt.xlabel('Race')
plt.ylabel('Nombre de personnes')
plt.title('Répartition des races par niveau d\'éducation')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Niveau d\'éducation')
plt.tight_layout()
plt.show()
```
